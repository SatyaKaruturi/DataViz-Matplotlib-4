# Mission 1

Crée un scatter plot entre deux variables x et y de ton choix. Ajoute une légende et met les valeurs dans une couleur autre que noir.

```
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(20)
y = np.random.randn(20)


colors = np.random.rand(20)

lo = plt.scatter(x, y, c=colors, marker= 'o', alpha=0.5)
llo = plt.scatter(x, y, c=colors, marker= 'x', alpha=0.5)

plt.legend((lo, llo),
           ('Low Outlier', 'LoLo'),
           scatterpoints=1,
           loc='upper right',
           ncol=2,
           fontsize=9)

plt.show()
```

# Mission 2

Commente le code ci-dessous

```
import numpy as np  
import matplotlib.pyplot as plt #Import matplotlib pour crée les plots


rng = np.random.RandomState(0) #expose un certain nombre de méthodes pour générer des nombres aléatoires à partir d'une variété de distributions de probabilités
x = rng.randn(100) #crée un tableau de forme spécifiée et le remplit de valeurs aléatoires conformément à la distribution normale standard
y = rng.randn(100)
colors = rng.rand(100) #crée les colors pour l'array de random numbers
sizes = 1000 * rng.rand(100) #calucule de size 

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis') #crée le plot scatter avec les valuers x et y avec colors, sizes aprés confirm la density de visibilty avec alpha, enplus donné le colourmap
                            #L'objet viridis est un objet appelable qui, lorsqu'il est transmis à un float compris entre 0 et 1, renvoie une valeur RGBA à partir de la palette de couleurs
plt.colorbar(); #pour montre un legend de colors
```

# Mission 3

Modifie le code ci-dessous pour que :
- Les valeurs du scatter soient issues pour l'axe des x du cosinus, et pour l'axe y du sinus pour des valeurs allant de 0 à 50

```
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(19680801)

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots()
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set(xlabel='entry a', ylabel='entry b')
plt.show()
```
